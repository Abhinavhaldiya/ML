```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Create a dataset
from sklearn.datasets import make_classification

X,y=make_classification(n_samples=1000,n_classes=2,n_features=10,random_state=15)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
X_train
```

```python
from sklearn.linear_model import LogisticRegression

logistic=LogisticRegression()
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
y_pred
```

```python
## Peformance Metrics
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_true=y_test,y_pred=y_pred)
score=accuracy_score(y_test,y_pred)
print(score)
print(cm)
print(classification_report(y_test,y_pred))
```

## Hyperparameter tuning and Cross Validation


```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
#Grid Search Cv
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold(n_splits=5)
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
## Peformance Metrics
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_true=y_test,y_pred=y_pred)
score=accuracy_score(y_test,y_pred)
print(score)
print(cm)
print(classification_report(y_test,y_pred))
```

```python
## Randomized CV
from sklearn.model_selection import RandomizedSearchCV

model=LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred=randomcv.predict(X_test)
```

```python
## Peformance Metrics
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_true=y_test,y_pred=y_pred)
score=accuracy_score(y_test,y_pred)
print(score)
print(cm)
print(classification_report(y_test,y_pred))
```

## Logistic Regression with ROC curve and ROC AUC score

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve
import matplotlib.pyplot as pyplot
from sklearn.metrics import roc_auc_score
%matplotlib inline
```

```python
X,y=make_classification(n_samples=1000,n_classes=2,random_state=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
## Create a dummy model with default 0 as output(majority class)
dummy_model_prob=[0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
#Prediction based on Probability
model_prob=model.predict_proba(X_test)
```

```python
model_prob
```

```python
## Lets focus on the positive outcome
model_prob=model_prob[:,1]
```

```python
model_prob
```

```python
## Lets calculate the scores
dummy_model_auc=roc_auc_score(y_test,dummy_model_prob)
model_auc=roc_auc_score(y_test,model_prob)
print(dummy_model_auc)
print(model_auc)
```

```python
# Calculate ROC Curve
dummy_fpr,dummy_tpr,_=roc_curve(y_test,dummy_model_prob)
model_fpr,model_tpr,thresholds=roc_curve(y_test,model_prob)
```

```python
model_fpr,model_tpr
```

```python
thresholds
```

```python
##Plotting the dummy and model
pyplot.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy Model')
pyplot.plot(model_fpr,model_tpr,label='Logistic Model')

pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')

pyplot.legend()
pyplot.show()
```
